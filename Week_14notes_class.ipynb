    Design a game called guess number:

a)       Set a secret number.

b)      Ask the user to input a number

c)       Compare the user's number to the secret number you set up and print if the user's number is too high or too low or equal to the secret number.

```python
# sample code from Lisa
secretnumber=8 
secretnumberguess=input('What is your guess for the number?  ')
secretnumberguess=int(secretnumberguess) # change the data type from string to int

if secretnumberguess==secretnumber:
    print('you guessed the number!')
elif secretnumberguess<secretnumber:
    print ('too small')
else:    
    print ('too big')
```
